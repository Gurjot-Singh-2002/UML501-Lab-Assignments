```python
# Convert BirthDate to Age
df['BirthDate'] = pd.to_datetime(df['BirthDate'])
df['Age'] = 2024 - df['BirthDate'].dt.year
df = df.drop(columns=['BirthDate'])
```

```python
# Drop irrelevant columns
columns_to_drop = ['CustomerID', 'Title', 'FirstName', 'MiddleName', 'LastName', 'Suffix',
                   'AddressLine1', 'AddressLine2', 'StateProvinceName', 'CountryRegionName',
                   'PostalCode', 'PhoneNumber', 'LastUpdated']
df_cleaned = df.drop(columns=columns_to_drop)
```

# Part (c)

```python
df.dtypes
```

```python
ordinal = []
continous = []
categorical_string = []

for col, dtype in df.dtypes.items():
    if dtype==int:
        ordinal.append(col)
    elif dtype==float:

        continous.append(col)
    else:
        categorical_string.append(col)

print("Ordinal (rank based) features : \n{o}\n\n".format(o=ordinal))
print("Continous features : \n{c}\n\n".format(c=continous))
print("Categorical string features : \n{s}\n\n".format(s=categorical_string))
```

# Question 2

# Part(a)

```python
# Drop the columns that are mostly null
columns_to_drop = ['Title','Suffix', 'MiddleName', 'AddressLine2']
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

# Part(b)

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, KBinsDiscretizer, OneHotEncoder,StandardScaler,RobustScaler,MaxAbsScaler
```

```python
# Normalization
scaler = MinMaxScaler()
df['YearlyIncome_scaled'] = scaler.fit_transform(df[['YearlyIncome']])
df['Age_scaled'] = scaler.fit_transform(df[['Age']])
```

```python
df.head()
```

# Part(c)

```python
# Discretization (Binning)
discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')
df['YearlyIncome_binned'] = discretizer.fit_transform(df[['YearlyIncome_scaled']])
```

```python
df.head()
```

# Part (d)

```python
# Step 4: Standardization
scaler = StandardScaler()
df['YearlyIncome_scaled'] = scaler.fit_transform(df[['YearlyIncome']])
df['Age_scaled'] = scaler.fit_transform(df[['Age']].values.reshape(-1,1)).flatten()
```

```python
df.head()
```

# Part (e)

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_data = encoder.fit_transform(df[['Gender', 'MaritalStatus', 'Education', 'Occupation', 'CountryRegionName', 'StateProvinceName']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Gender', 'MaritalStatus', 'Education', 'Occupation', 'CountryRegionName', 'StateProvinceName']))

label_encoder = LabelEncoder()
df['Gender_encoded'] = label_encoder.fit_transform(df['Gender'])


final_df = pd.concat([df, encoded_df], axis=1)
final_df.drop(['Gender', 'MaritalStatus', 'Education', 'Occupation', 'CountryRegionName', 'StateProvinceName'], axis=1, inplace=True)

print("\nFirst 5 rows of the final dataframe:")
print(final_df.head())
```
